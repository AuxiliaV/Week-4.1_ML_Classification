```python
import pandas as pd
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset=dataset.drop("User ID",axis=1)
dataset.columns
```

```python
dataset=pd.get_dummies(dataset,drop_first=True).astype(int)
dataset
```

```python
indep=dataset[['Age','EstimatedSalary','Gender_Male']]
indep.shape
```

```python
dep=dataset[['Purchased']]
dep.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(indep,dep,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
para={'C':[0.01,0.1,1.0,10.0],'solver':['lbfgs','liblinear','newton-cg','newton-cholesky','sag','saga']}
grid=GridSearchCV(LogisticRegression(),para,refit=True,verbose=3,n_jobs=1,scoring='f1_weighted')
grid.fit(x_train,y_train)
```

```python
y_pred=grid.predict(x_test)
```

```python
re=grid.cv_results_
table=pd.DataFrame.from_dict(re)
table
```

```python
table.to_excel('Grid_LogisticRegression_results.xlsx',index=False)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```

```python
from sklearn.metrics import f1_score
F1_score=f1_score(y_test,y_pred,average='weighted')
```

```python
print("the best parameter of logistic regression is: {}".format(grid.best_params_),F1_score)
```

```python
from sklearn.metrics import roc_auc_score
score=roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
score
```
